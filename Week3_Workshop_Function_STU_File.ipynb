`(Area = 3.14 √ó radius √ó radius)`

**Output:**
```

Larger circle area: 314.0
Combined area: 78.26

````python
# Solution:
import math
def circle_area(radius):
    area = math.pi * radius * radius
    return area

area1 = circle_area(radius1)
area2 = circle_area(radius2)

print(f"Circle area: {area1}")
print(f"Larger circle area: {area2}")
print(f"Combined area: {area1 + area2}")
````

````python
# Solution:
import math
def circle_area(radius):
    area = math.pi * radius * radius
    return area

radius1 = int(input("Enter radius of first circle: "))
radius2 = int(input("Enter radius of second circle: "))
print("=" * 30)
print("üü¢ Area of the Circle: üü¢")
print("=" * 30)

area1 = circle_area(radius1)
area2 = circle_area(radius2)

print(f"Circle area: {area1}")
print(f"Larger circle area: {area2}")
print(f"Combined area: {area1 + area2}")
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:
def total_and_average(score1, score2, score3):
    total = score1 + score2 + score3
    average = total / 3
    return total, average

score1 = float(input("Enter first game score: "))
score2 = float(input("Enter second game score: "))
score3 = float(input("Enter third game score: "))

total, average = total_and_average(score1, score2, score3)
print(f"Total: {total}, Average: {average}")
print(f"Your total is {total} and average is {average}")
````

````python
# Solution
def total_and_average(score1, score2, score3):
    total = score1 + score2 + score3
    average = total / 3
    return total, int(average)

score1 = int(input("Enter first game score: "))
score2 = int(input("Enter second game score: "))
score3 = int(input("Enter third game score: "))

print("=" * 30)
print("‚ú® Total & Average from Three Game Scores: ‚ú®")
print("=" * 30)

total, average = total_and_average(score1, score2, score3)
print(f"Total: {total}")
print(f"Average: {average}")
print(f"Your Total is: {total}")
print(f"Your Average is: {average}")
````
